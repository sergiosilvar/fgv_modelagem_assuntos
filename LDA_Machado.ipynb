Como o interesse da análise é uma modelagem dos tópicos das obras do próprio autor, iremos remover da lista de arquivos os identificadores para as traduções e miscelâneas.



```python
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)
    
fileids = [id for id in fileids if id[:id.index('/')] not in ['miscelanea','traducao']]
fd = nltk.FreqDist([id[:id.index('/')] for id in fileids])
ListTable(fd.items())
```

A primeira etapa do trabalho consiste em remover dos documentos palavras indesejadas, de pouco valor semântico ao objetivo da modelagem, tais como preposições e artigos. Essas palavras são conhecidas como [*stop words*](http://en.wikipedia.org/wiki/Stop_words). A princípio, utilizaremos os stop words disponívies na função `stopwords(language)` do NLTK:

Iremos construir um corpus estruturado em lista de palavras para nossa análise a partir do corpus do NLTK. Já podemos aproveitar o processo de transformar o texto em tokens para remover palavras desnecessáiras à análise:

1. stop words;
1. palavras  menores que 3 caracteres;
1. palavras que ocorrem menos de 10 vezes.

Iremos armazenr os documentos processados em arquivos, para facilitar as utilizações futuras.

```python
stopwords = nltk.corpus.stopwords.words('portuguese')

print ', '.join(stopwords)
```

#### Aplicação do LDA com GENSIM
Utilizamos o GENSIM para a aplicação do LDA às obras de Machado de Assis. Construímos os métodos abaixo para auxiliar no uso da biblioteca.

* `executa_lda` - 
* `topicos`

Para executar o LDA com o Gensim, ao contrário do que está [na documentação](http://radimrehurek.com/gensim/tut1.html#from-strings-to-vectors), primeiro é necessário construir a estrutura que representa o vocabulário a ser analisado, para após construir o [*bag-of-words*](http://en.wikipedia.org/wiki/Bag_of_words). Nessa primeira etapa, carregamos todo o corpus em memória, e o passamos como parâmetro do construtor da classe [Dictionary](http://radimrehurek.com/gensim/corpora/dictionary.html):



```python
def executa_lda(corpus, K):

    # Marca o início do LDA.
    t0 = datetime.datetime.now()

    # Cria dicionário Gensim com vocabulário a partir do corpus.
    sets = [set(c) for c in corpus]
    vocab = set.union(*sets)
    dictionary = corpora.Dictionary([vocab])

    # Cria o bag-of-words.
    bag = [dictionary.doc2bow(doc) for doc in corpus]

    lda = ldamodel.LdaModel([bag[0]], id2word=dictionary, num_topics=K )

    for i in range(1,len(bag)):
        tam = len(bag)
        print 'Atualizar LDA com o documento {}/{}. Total {:.2%}'.format(i, tam, 1.0*(i+1)/tam)
        lda.update([bag[i]])

    delta_t = datetime.datetime.now()-t0
    print 'LDA dos {} documentos terminado em em {} segundos.'.format(len(bag), delta_t.seconds) # Leva 285 segundos.
    return lda

def topicos(lda, n_words=10):       
    top,dist = _obtem_topicos( lda, n_words=n_words)
    t = PrettyTable()
    t._repr_html_ = t.get_html_string
    t.padding_width = 1
    for i in range(len(top)):
        t.add_column('Topico{}'.format(i+1), top[i])

    return t
    
def distribuicoes(lda, n_words=10):       
    top,dist = _obtem_topicos( lda,  n_words=n_words)
    t = PrettyTable()
    t._repr_html_ = t.get_html_string
    t.padding_width = 1
    for i in range(len(top)):
        t.add_column('Topico{}'.format(i+1), dist[i])

    return t
    
        
def _obtem_topicos(lda, n_words=10):
    topics = []
    dist = []
    for t in _gensim_extract(lda, n_words=n_words):
        words = []
        prob = []
        for item in t:
            words.append(item[0])
            prob.append(item[1])
        topics.append(words)
        dist.append(prob)
    return topics,dist


def _gensim_extract(lda, n_words=10):
    p = []
    for top in lda.show_topics(topics=lda.num_topics, topn=n_words):
        lista_p = top.split(' + ') 
        vk = [(l.split('*')[1],float(l.split('*')[0])) for l in lista_p ]
        p.append(vk)
        #[l[1] for l in lista_v]
    return p
```
