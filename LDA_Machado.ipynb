### Pré processamento do corpus
Vamos verificar como é o corpus do escritor Machado de Assis presente na biblioteca NLTK.

```python
import nltk, datetime, os
from prettytable import PrettyTable
from gensim import corpora
from gensim.models import ldamodel
from IPython.display import HTML
from ListTable import ListTable
from nltk.stem import RSLPStemmer
import sys
reload(sys)
sys.setdefaultencoding('utf-8')
```

```python
# Identificadores dos documentos do corpus.
machado = nltk.corpus.machado
print machado.readme().encode('utf-8') # Caso não apareça aqui ver saída na tela de comando.

```

```python
fileids = machado.fileids()
fd = nltk.FreqDist([id_[:id_.index('/')] for id_ in fileids])
ListTable(fd.items())
```

Podemos ver que o corpus das obras  contém uma coleção de 137 contos, 45 críticas, 24 crônicas, 7 poesias, 10 peças de teatro, 10 romances, 10 miscelâneas, 3 traduções, totalizando 246 documentos. 

Como o interesse da análise é uma modelagem dos tópicos das obras do próprio autor, iremos remover da lista de arquivos os identificadores para as traduções e miscelâneas.



```python
fileids = [id for id in fileids if id[:id.index('/')] not in ['miscelanea','traducao']]
fd = nltk.FreqDist([id[:id.index('/')] for id in fileids])
ListTable(fd.items())
```

A primeira etapa do trabalho consiste em remover dos documentos palavras indesejadas, de pouco valor semântico ao objetivo da modelagem, tais como preposições e artigos. Essas palavras são conhecidas como [*stop words*](http://en.wikipedia.org/wiki/Stop_words). A princípio, utilizaremos os stop words disponívies na função `stopwords(language)` do NLTK:

```python
# Lista de stopwords em português do NLTK.
stopwords = nltk.corpus.stopwords.words('portuguese')

print ', '.join(stopwords)
```

Iremos construir um corpus estruturado em lista de palavras para nossa análise a partir do corpus do NLTK. Já podemos aproveitar o processo de transformar o texto em tokens para remover palavras desnecessáiras à análise:

1. stop words;
1. palavras  menores que 3 caracteres;
1. palavras que ocorrem menos de 10 vezes.



```python
# Marca o início da tokenização.
t0 = datetime.datetime.now()

# Armazena o tipo de documento para referência futura.
indice = []

# Vocabulário.
vocab = []

# Documentos processados.
corpus = []

for fileid in fileids:
    t1 = datetime.datetime.now()

    print 'Processando %s...' % fileid 
    indice.append(fileid[:fileid.index('/')])
    
    # Carrega todas as palavras.
    words = machado.words(fileid)

    # Filtra palavras menores que 3 caracteres, stopwords, e que ocorrem menos de 10 vezes.
    fd = nltk.FreqDist(words)
    doc = [w.lower() for w in words if len(w) > 3 and  w.lower() not in stopwords and fd[w] >= 10]
    
    # Atualiza vocabulário.
    vocab = vocab + fd.keys()
    
    # Adiciona ao corpus.
    corpus.append(doc)

    # Salva o documento processado em um novo arquivo.
#    if not os.path.isdir('./machado/tokens/'): os.makedirs('./machado/tokens/')
#    f = open('./machado/tokens/'+fileid.replace('/','_'),'w')
#    for w in doc:
#        f.write(w+u'\n')
#    f.close()
    
    delta_t = datetime.datetime.now()-t1
    print '\'%s\' processado em %d segundos.'%(fileid, delta_t.seconds) 
    
delta_t = datetime.datetime.now()-t0
print 'Processamento dos %d documentos terminado em em %d segundos.'%(len(fileids), delta_t.seconds) # Leva 177 segundos.
```

#### Aplicação do LDA com GENSIM
Utilizamos o GENSIM para a aplicação do LDA às obras de Machado de Assis. 

Conforme [a documentação](http://radimrehurek.com/gensim/tut1.html#from-strings-to-vectors), primeiro é necessário construir o vocabulário do corpus, de acordo com sua estrutura, que atribui um identificador inteiro para cada palavra do vocabulário. Para isso, basta passar o corpus como parâmetro do construtor da estrutura que armazenará o dicionário, por exemplo:

    dictionary = gensim.corpora.Dictionary([vocab])

Daí então podemos construir construir o [*bag-of-words*](http://en.wikipedia.org/wiki/Bag_of_words), estrutura que representa as plavras e o número de ocorrência de cada palavra por documento:

    bag = [dictionary.doc2bow(doc) for doc in corpus]
    print bag
    >> [[(0, 1), (1, 1), (2, 1)],
       [(0, 1), (3, 1), (4, 1), (5, 1), (6, 1), (7, 1)],
       [(2, 1), (5, 1), (7, 1), (8, 1)],
       [(1, 1), (5, 2), (8, 1)],
       [(3, 1), (6, 1), (7, 1)]]

O exemplo acima mostra um corpus de 5 documentos. O primeiro documento tem 3 palavras onde cada palavra é representada por uma tupla onde o primeiro elemento da tupla é o identificador numérico da palavra no vocabulário e o segundo elemento é  a quantidade de ocorrências dessa palavra no documento. Da mesma forma, o segundo documento tem 6 palavras, o terceiro tem 4 palavras, e assim por diante. 

Após a construção do bag-of-words podemos executar o LDA chamando o método `gensim.models.ldamodel.LdaModel` passando como parâmetros o bag-of-words, o dicionário com o vocabulário, e o número de tópicos:

        lda = ldamodel.LdaModel(bag, id2word=dictionary, num_topics=K )

Aa variável `lda` armazena o resultado do modelo. Para visualizar os tópicos, recorremos à funçao `show_topics()`, que aceita como parâmetros o número de topicos a exibir, sendo o máximo o número de tópico definido anteriormente, e o número de palavras mais relevantes a exbir:
    
    lda.show_topics(3,5)
    >> ['0.165*alves + 0.036*merc\xc3\xbario + 0.036*j\xc3\xbapiter + 0.033*marte + 0.032*rubi\xc3\xa3o',
     '0.079*homens + 0.053*cena + 0.051*cidade + 0.048*elisa + 0.025*pinheiro',
     '0.173*todas + 0.140*sobre + 0.100*rosto + 0.093*terra + 0.032*raquel']

No exemplo acima exibimos as 5 palavras mais relevantes para os 3 primeiros tópicos. Nota-se que o resultado precisa ser processado para uma melhor visualização. Até a conclusão deste relatório não encontramos uma estrutura da própria biblioteca para apresentar os tópicos e as distribuições de probabilidades de forma mais legível, então recorremos então à construção das funções `topicos` e `distribuições`:

* `topicos` - visualização textual dos tópicos encontrados pelo LDA do GENSIM.
* `distribuicoes`- visualização textual da distribuição de probabilidade das palavras por tópicos encontrados pelo LDA do GENSIM.

Ambas as funções retornam uma tabela HTML que integra-se automaticamente com o IPython Notebook e é rederizado normalmente pelo mesmo. 

O exemplo acima demontra a execução do modelo LDA passando todo o corpus de uma única vez. Há uma outra abordagem chamada online LDA, que ajusta as distribuições das palavras para novos documentos, sem precisar recorrer aos documentos anteriores, tomando como base a distribuição anterior. Construímos os métodos abaixo para auxiliar no uso da biblioteca para estes dois modos:

* `lda_batch` - executa a implementação do LDA da biblioteca GENSIM com o corpus e número de tópicos informado, atualizando o modelo com todos os documentos do corpus de uma só vez.
* `lda_online` - executa a implementação do LDA da biblioteca GENSIM com o corpus e número de tópicos informado, atualizando o modelo com a iteração sobre cada docuemnto do corpus.

A seguir, apresentamos as funções construídas para a execução do LDA.


#### Funções para a execução do LDA.

```python

def lda_batch(corpus, K):
    ''' 
    Executa o LDA em modo batch, em que todos os docuemntos do corpus são avaliados de uma única vez.
    '''
    # Marca o início do LDA.
    t0 = datetime.datetime.now()

    # Cria dicionário Gensim com vocabulário a partir do corpus.
    sets = [set(c) for c in corpus]
    vocab = set.union(*sets)
    dictionary = corpora.Dictionary([vocab])

    # Cria o bag-of-words.
    bag = [dictionary.doc2bow(doc) for doc in corpus]

    lda = ldamodel.LdaModel(bag, id2word=dictionary, num_topics=K )

    delta_t = datetime.datetime.now()-t0
    print 'LDA dos {} documentos terminado em em {} segundos.'.format(len(bag), delta_t.seconds) # Leva 285 segundos.
    return lda

def lda_online(corpus, K):
    '''
    Executa o LDA em modo online, onde os documentos do corpus são avaliados iterativamente.
    '''

    # Marca o início do LDA.
    t0 = datetime.datetime.now()

    # Cria dicionário Gensim com vocabulário a partir do corpus.
    sets = [set(c) for c in corpus]
    vocab = set.union(*sets)
    dictionary = corpora.Dictionary([vocab])

    # Cria o bag-of-words.
    bag = [dictionary.doc2bow(doc) for doc in corpus]

    lda = ldamodel.LdaModel([bag[0]], id2word=dictionary, num_topics=K )

    for i in range(1,len(bag)):
        tam = len(bag)
        print 'Atualizar LDA com o documento {}/{}. Total {:.2%}'.format(i, tam, 1.0*(i+1)/tam)
        lda.update([bag[i]])

    delta_t = datetime.datetime.now()-t0
    print 'LDA dos {} documentos terminado em em {} segundos.'.format(len(bag), delta_t.seconds) # Leva 285 segundos.
    return lda

def topicos(lda, n_words=10):
    '''
    Representação textual dos tópicos encontrados pelo LDA do GENSIM.
    '''
    top,dist = _obtem_topicos( lda, n_words=n_words)
    t = PrettyTable()
    t._repr_html_ = t.get_html_string
    t.padding_width = 1
    for i in range(len(top)):
        t.add_column('Tp{}'.format(i+1), top[i])

    return t
    
def distribuicoes(lda, n_words=10):
    '''
    Representação textual da distribuição de probabilidade das palavras por tópicos encontrados pelo LDA do GENSIM.
    '''
    
    top,dist = _obtem_topicos( lda,  n_words=n_words)
    t = PrettyTable()
    t._repr_html_ = t.get_html_string
    t.padding_width = 1
    for i in range(len(top)):
        t.add_column('Tp{}'.format(i+1), dist[i])

    return t
    
        
def _obtem_topicos(lda, n_words=10):
    topics = []
    dist = []
    for t in _gensim_extract(lda, n_words=n_words):
        words = []
        prob = []
        for item in t:
            words.append(item[0])
            prob.append(item[1])
        topics.append(words)
        dist.append(prob)
    return topics,dist


def _gensim_extract(lda, n_words=10):
    p = []
    for top in lda.show_topics(topics=lda.num_topics, topn=n_words):
        lista_p = top.split(' + ') 
        vk = [(l.split('*')[1],float(l.split('*')[0])) for l in lista_p ]
        p.append(vk)
        #[l[1] for l in lista_v]
    return p
```

### Estudo da Qualidade da Modelagem

Notamos que a execução no modo online apresenta tópicos com discernimento mais apurado que o modo batch. No modo batch, muitas palavras se repetem entre os tópicos, e a distribuição de probabilidades das palavras assumem um valor bem pequeno quando comparado com a mesma distribuição no modo online. Entretanto, para avaliar objetivamente a qualidade da modelagem, devemos determinar melhor o que desejamos obter do modelo. Nesta primeira aplicação apenas exercitamos o uso do GENSIM para todo o corpus das obras de Machado de Assis mas, essa coleção, contém documentos de naturezas diferentes, como crônicas e peças de teatro, o que pode misturar tópicos distintos e invalida a premissa do LDA de que todos os documentos do corpus foram gerados a partir dos mesmos tópicos.

Propomos então analisar a coleção de contos, que são os documentos em maior número disponível nesse corpus. 



```python
idx = [i for i,x in enumerate(indice) if x == 'contos']
sel_corpus = corpus[min(idx):max(idx)+1]

ldac = lda_online(sel_corpus, K=13)

```

```python
topicos(ldac,15)
```

```python
distribuicoes(ldac,15)
```

Observamos que os tópicos para esta coleção de contos apresentam-se bem definidos, as palavras mais relevantes, majoritariamente, são exclusivas aos seus tópicos, sendo que muitas delas são nomes dos personagens dos contos. Como trata-se de uma coleção literária, não é supresa notar que este seria o resultado esperado, pois à exceção dos stopwords que já foram removidos, os nomes dos personanges repetem-se ao longo dos textos. afinal, para que a história aconteça, algum personagem tem que desempenhar uma ação. De certa forma, este resultado valida a execução do LDA mas, a depender do objetivo da análise em vigência, os nomes dos personagens poderiam ser tratados como stopwords, a fim de permitir a definição mais semântica do tópico. Em termos numéricos, vemos que as dentre as as palavras mais relevantes, as probabilidades são bem distintas até a 5.a palavra, e a partir desta as probabilidades assumem pouca diferença entre si. O tópico 6, por exemplo, tem a partir da 8.a palavra mais relevante probabilidades próximo ao fim da precisão dada como saída da biblioteca GENSIM. Nota-se que as 7 palavras mais relevantes deste tópico em questão são nomes de personangens, então não temos uma expressão semântica par ao significado deste tópico.

### Considerações finais

A proposta deste trabalho era avaliar a aplicação do LDA para um corpus de cunho literário, com a expectativa de observar tópicos que mostrassem os temas relevantes às obras. Esperava-se que personagens e suas ações figurassem entre as palavras mais relevantes, e que se pudesse também identificar os elementos literários que definem o estilo de determinado autor. 

Utilizamos a biblioteca GENSIM, da linguagem Python, apara a aplicação do LDA, e verificamos que o resultado é apresentado num formato específico, mais para uma interpretação qualitativa do que quantitativa. Criamos métodos para analisar este resultado e obter as palavras mais relevantes e suas respectivas probabilidades de pertecerem ao tópico em questão. O formato de saída restringiu a aplicação de avaliações quantitativas sobre a distribuição de probabilidades das palavras.

Aplicamos o LDA ao corpus do escritor brasileiro Machado de Assis, disponível na biblioteca Python NLTK, e em uma primeira aborgadem, os tópicos resultantes de toda a sua obra não apresentam o resultado esperado. Propusemos avaliar somente a coleção de contos, para que a análise fosse feita dentro de um conjunto de mesmo cunho literário e escolhemos a coleção de contos para este fim. O resultado foi a observação ainda dos personagens como tópicos, não mais identificados por seus nomes próprios, mas por suas atuações, profissões e tratamos de manual e iterativamente removê-los deste corpus. Ao fim, chegamos a tópicos definidos por substantivos que descrevem a  atuação destes, e outros substantivos com funções semânticas diversas que restringem uma contextualização mais específica do tópico.

Desta forma, podemos considerar que, neste primeira tentativa, a utilização do LDA para a identificação de tópicos em obras literárias não foi bem sucedida. A escolha do corpus toma papel de destaque na qualidade da modelagem pois para que o LDA possa convergir a um resultado apropriado, é necessário uma quanitade de docuemnto razoavelmente grande. Não são muitos os autores capazes de escrever algumas centenas de docuemntos diferentes. Escolhemos o corpus do Machado de Assis por uma conveniência de grande parte de sua obra estar  disponível no NLTK, mas como resultado desta primeira aborgagem, uma análise de obras literárias merece especial atenção no objetivo a se alançar. Se a intenção é avaliar os tópicos presentes na obra de um único autor, como é o caso deste trabalho, uma proposta pode ser a segmentação dos documentos do corpus dentro de cada obra, por exemplo, tomando-se cada capítulo como um doucmento. Talvez seja mais plausível a avaliação de tópicos para um determinado estilo literário, Romantismo, Arcadismo, Realismo, utilizando de diversos autores, sendo a disponibilidade e questões legais de direito de uso fatores limitantes. Não obstante, podemos supor que qualquer que seja o escopo da análise, uma especial atenção deverá ser dada a como tratar a ocorrência das palavras que identificam personagens.

Como proposta de trabalho futuro podemos destacar a estudo da API do GENSIM para observar configurações o funções que possam ser aplicados para refinar a execução do algoritmo ou interpretação dos resultados obtidos. Outro item é a avaliação de outros modelos para análises literárias.
